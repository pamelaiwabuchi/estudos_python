```python
#not using list comprehension

values = []
for x in range(10):
    values.append(x)
    print(values)
```

```python
#using list comprehension
values = [x+1 for x in range(10)]
print(values)
```

```python
#Comprehension Condition
evens = []
for number in range(50):
    is_even = number % 2 == 0
    if is_even:
        evens.append(number)
    print (evens)
```

```python
#Comprehension Condition
evens = [number for number in range(50) if number % 2 == 0]
print(evens)
```

```python
#Not using comprehesion with multiple conditions
#strings that start with 'a' and end in 'y'
options = ["any", 'albany', 'apple', 'world', 'hello', ""]
valid_strings = []
for string in options:
    if len(string) <=1:
        continue
    if string[0] != 'a':
        continue
    if string[-1] != 'y':
        continue

    valid_strings.append(string)
    print(valid_strings)
```

```python
#Using comprehesion with multiple conditions
options = ["any", 'albany', 'apple', 'world', 'hello', ""]
valid_strings = [
    string 
    for string in options 
    if len(string) >= 2 
    if string[0] == 'a' 
    if string[-1] == 'y'
]
print(valid_strings)
```

```python
#no multiple list comprehension
#flattening a matrix (list of lists)
matrix = [[1,2,3], [4,5,6], [7,8,9]]
flattened = []

for row in matrix:
    for i in row:
        flattened.append(i)
        print(flattened)
```

```python
#with multiple list comprehension
matrix = [[1,2,3], [4,5,6], [7,8,9]]
flattened = [num for row in matrix for num in row]
print(flattened)
```
