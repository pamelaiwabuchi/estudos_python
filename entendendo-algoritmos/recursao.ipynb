## Exemplo de Recursão - quando uma função chama a si mesma. - Capítulo 3

```python
def fat(x):
    if x ==1:
        return 1
    else: 
        return x * fat (x-1)
fat(3)
```

## Quicksort - algoritmo de ordenacao elegante

```python
def mdc(a: int, b: int) -> int:
    """
    Calcula o MDC usando o Algoritmo de Euclides.
    Seguindo o padrão de 'Dividir para Conquistar'.
    """
    print(f"Chamada: mdc({a}, {b})") # Para você ver a pilha em ação
    
    # Caso Base: se o resto for zero, o divisor 'a' é o MDC
    if b == 0:
        return a
        
    # Chamada Recursiva: o 'b' vira o novo 'a', e o resto vira o novo 'b'
    return mdc(b, a % b)

# Teste com os números 270 e 192
resultado = mdc(270, 192)
print(f"\nO MDC final é: {resultado}")
```
